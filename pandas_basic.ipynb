# Pandas DataFrame Reference – Examples

This notebook contains example usage of common `pandas.DataFrame` functions
based on the official pandas DataFrame API reference.

```python
import pandas as pd
import numpy as np
```

## Create DataFrame

```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': ['x', 'y', 'z']
})
df
```

## DataFrame Attributes

```python
df.shape
df.columns
df.index
df.dtypes
```

## Selection & Indexing

```python
df['A']
df.loc[1]
df.iloc[2]
df.at[0, 'B']
df.iat[1, 2]
```

## Adding / Modifying Columns

```python
df['D'] = df['A'] + df['B']
df.assign(E=df['A'] * 10)
```

## Statistics & Aggregations

```python
df.sum()
df.mean(numeric_only=True)
df.describe()
```

## Apply & Map

```python
df.apply(lambda col: col, axis=0)
df.applymap(str)
```

## Sorting

```python
df.sort_values(by='D', ascending=False)
df.sort_index()
```

## Missing Values

```python
df2 = pd.DataFrame({'A': [1, np.nan, 3], 'B': [4, 5, np.nan]})
df2.isna()
df2.fillna(0)
df2.dropna()
```

## GroupBy

```python
df3 = pd.DataFrame({'key': ['A', 'A', 'B'], 'value': [10, 20, 30]})
df3.groupby('key').sum()
```

## Merge & Join

```python
left = pd.DataFrame({'id': [1, 2], 'val': [100, 200]})
right = pd.DataFrame({'id': [2, 3], 'val2': [300, 400]})
pd.merge(left, right, on='id', how='outer')
```

## Window Functions

```python
df4 = pd.DataFrame({'num': [10, 20, 30, 40]})
df4.rolling(2).sum()
df4.expanding().mean()
```

## Export

```python
df.to_csv('dataframe_example.csv', index=False)
'CSV file saved'
```
